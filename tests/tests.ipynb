```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="Lorbus/Qwen3.6-27B-int4-AutoRound",
    base_url="http://127.0.0.1:8000/v1",   # ← your local proxy
    openai_api_key="dummy",                  # ← vLLM doesn't need a real key
    timeout=300,                             # ← long timeout for big model generation
    max_retries=1,
)

result = llm.invoke("write a story of around 100 words")

print(result.content)
```
